##5. What is mutability in Python?
    Mutability in Python is the property of an object that determines whether its value can be changed after creation.
    Mutable objects can be modified(Ex: list,dictionary,set)




##6. Why are lists mutable, but tuples are immutable?
    Lists are mutable because they are designed to allow modification of elements,such as adding,removing,or changing items,making them flexible for dynamic data.
    Tuples are immutable because they are meant to store fixed,constant data; once created,their values cannot be changed,which ensures data integrity and allows them to be used as dictionary keys.

##7. What is the difference between “==” and “is” operators in Python?
    

*   ==(Equality operator): Checks whether two objects have the same value/content.
*   is(Identity operator): Checks whether two variables refer to the same object in memory.






```python
a=[1,2,3]
b=[1,2,3]
print(a==b)  #True(same values)
print(a is b)  #False(different objects)
```

##8. What are logical operators in Python?

    The main logical operators are:

##9. What is type casting in Python?
     Type casting in Python is the process of converting one data type into another.
     Python provides built-in functions for this,such as:


*   int()- converts to integer
*   float()- converts to float
*   str()- converts to string



##10. What is the difference between implicit and explicit type casting?
    The difference between implicit and explicit type casting in Python is:
    1. Implicit type casting: Python automatically converts one data type to another during on operation to avoid data loss.
    Example: int to float.
    2. Explicit type casting: The programmer manually converts a data type using functions like int(),float(),or str().

##11. What is the purpose of conditional statements in Python?
    The purpose of conditional statements in Python is to make decisions based on given conditions.
    1. They allow a program to execute different blocks of code depending on whether a condition is true or false.
    2. They help control the flow of a program.
    3. They are used to implement logic and decision-making in programs.

##12. How does the elif statement work?
    The elif statement in Python is used to check multiple conditions in sequence.
    1. Python first evaluates the if condition.
    2. If the if condition is false,it checks the elif condition.
    3. The block of code under the first true condition is executed,and the rest are skipped.

##13. What is the difference between for and while loops?
    The difference between for and while loops in Python is:
    1. for loop: Used when the number of iterations is known in advance.It iterates over a sequence like a list,tuple or range.
    2. while loop: Used when the number of iterations is not known and the loop runs as long as a condition is true.
     

##14. Describe a scenario where a while loop is more suitable than a for loop.
    A while loop is more suitable than a for loop when the number of iterations is not known in advance.
    Scenario: When taking user input repeatedly until a correct value is entered,the loop must continue based on a condition,not a fixed count.

##13.  Write a Python program to find the sum of all even numbers between 1 and 50.

```python
sum=0
for i in range(1,50):
  if i%2==0:
    sum+=i
print("Sum of all even numbers between 1 and 50 is",sum)
```

##14. Write a program to reverse a string using a while loop.

```python
a=input("Enter a string: ")
i=len(a)-1
while i>=0:
  print(a[i],end="")
  i-=1
```

##15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
a=input("Enter a number: ")
fact=1
i=1
while i<=int(a):
  fact*=i
  i+=1
print("Factorial of",a,"is",fact)
```
